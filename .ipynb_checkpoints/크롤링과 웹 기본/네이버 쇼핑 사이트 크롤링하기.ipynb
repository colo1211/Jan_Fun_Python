```python
import requests 
from bs4 import BeautifulSoup 
```

```python
res= requests.get('https://search.shopping.naver.com/best100v2/main.nhn')
soup = BeautifulSoup (res.content,'html.parser')
data = soup.select('#popular_srch_lst > li > span.txt > a')
for index, i in enumerate(data): 
    print (index+1,"위.", i.get_text())
```

```python
res= requests.get('https://search.shopping.naver.com/best100v2/detail.nhn?catId=50000000&listType=B10002')
soup = BeautifulSoup (res.content,'html.parser')
data = soup.select('#productListArea > ul > li > p > a')
for index,i in enumerate(data):
    print (index+1,"위",i.get_text())
```
