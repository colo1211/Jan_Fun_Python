# 데이터 구조 JSON vs XML

## 다양한 데이터 읽기 - XML

* XML(Extensible Markup Language)
  - 특정 목적에 따라 데이터를 태그로 감싸서 마크업하는 범용적인 포멧
  - 마크업 언어는 태그 등을 이용하여 데이터의 구조를 기술하는 언어의 한 가지
  - 가장 친숙한 마크업 언어가 HTML
  - XML은 HTML과 마찬가지로 데이터를 계층 구조로 표현
  - XML 기본 구조
  <pre>
  <태그 속성="속성값">내용</태그>
  </pre>

    - 태그와 속성은 특정 목적에 따라 임의로 이름을 정해서 사용, 즉, 정해진 규칙(문법)에 따라서 작성하는 것이 아닌 내 마음대로 태그와 속성을 정해서 사용하는 방법

    - 다른 요소와 그룹으로 묶을 수도 있음

```python
import requests
from bs4 import BeautifulSoup  
import openpyxl


excel_file=openpyxl.Workbook()
excel_sheet = excel_file.active
excel_sheet.append(['번호','지역명','시간','이산화탄소'])
service_key='pD7%2Fw7r2g0ca%2FNunCwcYvN3kR2FtOvMtvodnRa1Tzipoy33ooQq%2B1nqCCbb%2B70u2sVZHwv9bRL8JT%2BzrCkQ4QQ%3D%3D'
params='&numOfRows=10&pageNo=1&sidoName=인천&ver=1.3' # 추가적인 요청변수를 적어준다. 

openapi_url='http://openapi.airkorea.or.kr/openapi/services/rest/ArpltnInforInqireSvc/getCtprvnRltmMesureDnsty?ServiceKey='+service_key+params

res=requests.get(openapi_url)
soup = BeautifulSoup(res.content,'html.parser')
data = soup.find_all('item') #css가 안쓰이기 때문에 select는 쓰지 못함
for index, i in enumerate(data) : 
    city_name=i.find('stationname')
    data = i.find('datatime')
    co2 = i.find('covalue')
    excel_sheet.append([index+1,city_name.get_text(),data.get_text(),co2.get_text()])
excel_file.save('인천시 미세먼지.xls')
excel_file.close()
```
